```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.feature_extraction.text import CountVectorizer
```

# 1. Load Data

```python
df = pd.read_csv('Pokemon.csv')
df.info()
df.shape
```

```python
# cek null
df.isnull().sum()
df.tail(15)
```

# 2. Feature Selection
Ada 3 fitur yang ingin dipakai sebaga parameter sistem rekomendasi, yaitu
requirement feature :
- Type 1
- Generation
- Legendary

```python
df_selection = df[['Type 1', 'Generation', 'Legendary']]
df_selection.head()
```

Ternyata, ada fitur yang mengandung Kategorical Data yaitu Fitur "Type 1". Untuk menangani hal ini maka akan dilakukan proses mapping value dari fitur ini.

```python
print(df_selection['Type 1'].nunique(), "tipe kategorical data")
df_selection['Type 1'].unique()
```

```python
# Mengekstraksi fitur Type 1

extractor = CountVectorizer()
print(extractor.fit(df_selection['Type 1']))

print("fitur kata",extractor.get_feature_names())
print("jumlah fitur kata : ",len(extractor.get_feature_names()))
```

```python
# Transformasi matrix dari data kategorical Type 1

fitur_name = extractor.get_feature_names()
trf_matrix = extractor.transform(df_selection['Type 1']).toarray()
fitur_name = extractor.get_feature_names()

# print(trf_matrix)

df_matrix = pd.DataFrame(trf_matrix, columns=fitur_name)
df_matrix 
```

```python
# menambahkan matriks dengan fitur yang lainnya (Generation and Legendary)
df_matrix['gen'] = df['Generation']
df_matrix['legend'] = list(map(lambda leg : 1 if leg else 0 , df['Legendary']))
df_matrix
```

# Hitung Angka Kemiripan Dengan Cosine Similarity

```python
# cosinus similarity
from sklearn.metrics.pairwise import cosine_similarity

# ini adalah matriks kemiripan masing masing konten pokemon
similarity_scores = cosine_similarity(trf_matrix)
print(similarity_scores.shape)

dfsimi = pd.DataFrame(similarity_scores)
dfsimi.to_csv('similarity.csv', sep=',', index=False)
# df.to_csv(file_name, sep='\t')
```
